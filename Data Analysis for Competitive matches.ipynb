## Data Preparation

We only need the following columns: bResult, rResult, blueTopChamp, blueJungleChamp, blueMiddleChamp, blueADCChamp, blueSupportChamp, redTopChamp, redJungleChamp, redMiddleChamp, redADCChamp, redSupportChamp.

For simplicity, we can discard the team (color) information and further condense the above columns into fewer columns as our topic only focuses on which team composition is the best (without regard to what the team is competing against):
* win (0 = loss, 1 = win)
* top
* middle
* jungle
* bottom/adc

```python
blue = competitive[['blueTopChamp', 'blueJungleChamp', 'blueMiddleChamp', 'blueADCChamp', 'blueSupportChamp', 'bResult']]
red = competitive[['redTopChamp', 'redJungleChamp', 'redMiddleChamp', 'redADCChamp', 'redSupportChamp', 'rResult']]
blue.columns = ('top', 'jungle', 'middle', 'bottom', 'support', 'win')
red.columns = ('top', 'jungle', 'middle', 'bottom', 'support', 'win')
competitive = pd.concat([red, blue], axis=0).reset_index(drop=True)
match_roles = competitive.copy()
competitive.tail()
```

```python
competitive.info()
```

Let us now erase the roles (top, jungle, middle, adc, support) to simplify the problem

We will also rearrange the champions so that different permutations are regarded as the same.

```python
competitive.columns = ('p1','p2','p3','p4','p5','win')

def rearrange(series):
    win = series['win']
    series = series.drop('win').sort_values()
    series.index = ('p1', 'p2', 'p3', 'p4', 'p5')
    series['win'] = win
    return series
    
competitive = competitive.transform(rearrange, axis=1)
competitive.head()
```

```python
competitive = competitive.sort_values(by=['p1', 'p2', 'p3', 'p4', 'p5'])
sorted_com = competitive.copy()
competitive.head()
```

```python
cgroup = competitive.groupby(['p1', 'p2', 'p3', 'p4', 'p5'])
cgroup.apply(len).value_counts()
```

### Data preparation - Good Pick score

This portion will deal with calculating the good pick score for each team composition. For more details, please look at *Good Pick Score v2.ipynb*.

```python
def win_rate(n_wins, n):
    return 2*(n_wins/n) - 1

def goodpickscore(n_wins, n):
    p = 0.2
    w = win_rate(n_wins, n)
    s = np.sign(w)
    if s == 0:
        s = 1
    return w * np.log(n + p) + s * np.tanh(n)

def calcGoodPickScorePerGroup(group):
    n = len(group)
    n_wins = group['win'].sum()
    return goodpickscore(n_wins, n)

c = cgroup.apply(calcGoodPickScorePerGroup)

def updateScore(series):
    score = c[series['p1'], series['p2'], series['p3'], series['p4'], series['p5']]
    series['win'] = score
    return series

competitive = competitive.apply(updateScore, axis=1).drop_duplicates()

# Show the distribution of the winning rates
competitive['win'].value_counts().sort_index()
```

There are 7424 teams that have a positive good pick score, while there are only 7190 teams that have a negative good pick score. Hence, we can tell from the results that the teams only lean slightly more towards winning. This makes sense because majority of teams were only seen once, i.e played only in one competitive match. Hence, there wouldn't be a clear majority formed yet.

### Data Preparation - Dummify

In this section, we will dummify the dataframe

```python
# Retrieve the latest champion list
champs = pd.read_json('data/champion-9.2.1.json')
champs = pd.Series([0] * len(champs.index), index=champs.index.map(lambda x: x.lower()))
champs.head()
```

```python
def dummify(row):
    t = champs.copy()
    t[row['p1'].lower()] = 1
    t[row['p2'].lower()] = 1
    t[row['p3'].lower()] = 1
    t[row['p4'].lower()] = 1
    t[row['p5'].lower()] = 1
    t['win'] = row['win']
    return t

competitive = competitive.apply(dummify, axis=1)

# Change columns to integer as they were parsed as floats
for i in competitive.columns:
    competitive[i] = pd.to_numeric(competitive[i], downcast="integer")

# Save to file for model
competitive.to_csv('data/competitive-matches-processed/dummified.csv', index=False)

competitive.head()
```

### Data Analysis - Frequency

In this section, we will determine which champion is commonly played with other champions

```python
freq = pd.DataFrame(np.zeros((len(champs.index),len(champs.index)), dtype=np.int32), index=champs.index, columns=champs.index)

for index, row in sorted_com.iterrows():
    for i in ['p1','p2','p3','p4','p5']:
        ind = row[i].lower()
        freq.loc[ind, row['p1'].lower()] += 1
        freq.loc[ind, row['p2'].lower()] += 1
        freq.loc[ind, row['p3'].lower()] += 1
        freq.loc[ind, row['p4'].lower()] += 1
        freq.loc[ind, row['p5'].lower()] += 1

pd.options.display.max_columns = None
pd.options.display.max_rows = None
freq.to_csv('data/competitive-matches-processed/frequency.csv')
freq
```

```python
c = 0
def select(series):
    global c
    f = series.iloc[c]
    c += 1
    return f

num_plays = freq.apply(select, axis=1).sort_values()

with mpl.rc_context(rc={'ytick.labelsize':'x-large','figure.figsize':(16,6),'axes.titlesize':'xx-large', 'axes.xmargin': 0.01}):
    plt.bar(num_plays.index, num_plays)
    plt.xticks(rotation='vertical')
    plt.tight_layout()
    plt.title('Number of plays per champion in competitive matches')
    plt.show()
```

From above, we can see that gragas, a supportive-tanky jungler is the most picked champion for competitive matches. This is in line with our expections. In fact, 3 out of the top 5 are junglers. This might show that there is very limited choice for junglers in a competitive match, same could go for supports too.

If we were to choose a team composition from this graph, it would be (from our own biases)
* maokai, gragas, karma, ashe, thresh
* gnar, graves, orianna, lucian, thresh

```python
with mpl.rc_context(rc={'ytick.labelsize':'x-large','figure.figsize':(16,6),'axes.titlesize':'xx-large', 'axes.xmargin': 0.01}):
    lowest_played = num_plays.head(15)
    plt.bar(lowest_played.index, lowest_played)
    plt.title('Top 15 least played champions in competitive matches')
    plt.show()
```

```python
with mpl.rc_context(rc={'ytick.labelsize':'large','figure.figsize':(16,10),'axes.titlesize':'xx-large','figure.subplot.hspace':0.8,'axes.xmargin': 0.01}):
    
    fig, axes = plt.subplots(3,2)
    fig.suptitle('Top 15 champions per role in competitive matches', fontsize=30)
    
    top_plays = match_roles['top'].value_counts().nlargest(15)
    axes[0,0].bar(top_plays.index, top_plays)
    axes[0,0].set_title("Top")

    jungle_plays = match_roles['jungle'].value_counts().nlargest(15)
    axes[0,1].bar(jungle_plays.index, jungle_plays)
    axes[0,1].set_title("Jungle")

    mid_plays = match_roles['middle'].value_counts().nlargest(15)
    axes[1,0].bar(mid_plays.index, mid_plays)
    axes[1,0].set_title("Middle")

    bottom_plays = match_roles['bottom'].value_counts().nlargest(15)
    axes[1,1].bar(bottom_plays.index, bottom_plays)
    axes[1,1].set_title("Bottom")

    support_plays = match_roles['support'].value_counts().nlargest(15)
    axes[2,0].bar(support_plays.index, support_plays)
    axes[2,0].set_title("Support")
    
    for ax in fig.axes:
        plt.sca(ax)
        plt.xticks(rotation=90)
```

Maokai is clearly the leading top for games, while reksai, gragas and elise are the leading junglers. Orianna and Viktor are leading mids while sivir, lucian and ashe are leading bottoms, braum, thresh, alistar and karma are leading supports.
