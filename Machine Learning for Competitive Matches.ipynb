```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.dummy import DummyRegressor
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import r2_score

competitive = pd.read_csv('data/competitive-matches-processed/dummified.csv')
competitive.head()
```

```python
X = competitive.iloc[:, 0:-1]
y = competitive['win']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(r2_score(y_test, SVR().fit(X_train, y_train).predict(X_test)))
print(r2_score(y_test, LinearRegression().fit(X_train, y_train).predict(X_test)))
print(r2_score(y_test, DecisionTreeRegressor().fit(X_train, y_train).predict(X_test)))
print(r2_score(y_test, GradientBoostingRegressor().fit(X_train, y_train).predict(X_test)))
print(r2_score(y_test, DummyRegressor().fit(X_train, y_train).predict(X_test)))
```

```python
# Retrieve the latest champion list
champs = pd.read_json('data/champion-9.2.1.json')
champs = pd.Series([0] * len(champs.index), index=champs.index.map(lambda x: x.lower()))

def dummify(row):
    t = champs.copy()
    t[row.iloc[0].lower()] = 1
    t[row.iloc[1].lower()] = 1
    t[row.iloc[2].lower()] = 1
    t[row.iloc[3].lower()] = 1
    t[row.iloc[4].lower()] = 1
    return t

def predict(champs):
    global model
    return model.predict([dummify(pd.Series(champs))])

print("Top 5 worst performers based on correlation against goodpickscore: %f" % predict(['drmundo', 'viktor', 'evelynn', 'kindred', 'karma']))
print("Top 5 best performers based on correlation against goodpickscore: %f" % predict(['malzahar','kalista','azir','zac','camille']))
```
